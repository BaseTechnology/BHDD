```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

dataset = []

with open("data.pkl","rb") as file:
    dataset = pickle.load(file)
            
trainDataset = dataset["trainDataset"]
testDataset = dataset["testDataset"]

numberOfClass = 10

trainX = np.array([list(td["image"].flatten() / 255) for td in trainDataset])
trainY = np.array([td["label"] for td in trainDataset])
testX = np.array([list(td["image"].flatten() / 255) for td in testDataset])
testY = np.array([td["label"] for td in testDataset])
```

```python
def draw_plots(data, lbl, str_title):
    plt.figure(figsize=(12,6))
    plt.scatter(data[:, 0],
                data[:, 1],
                c=lbl,
                edgecolor='none',
                alpha=0.5,
                cmap=plt.get_cmap('jet', 10),
                s=5)
    plt.colorbar()
    plt.title(str_title)
    plt.show()
```

## Visualize PCA (n=2) projection of BHDD dataset

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(trainX)
```

```python
draw_plots(pca_result, trainY, "PCA (n=2) projection of training images of BHDD dataset")
```

## Visualize LDA (n=2) projection of BHDD dataset

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=2)  
lda_result = lda.fit_transform(trainX, trainY)  
```

```python
draw_plots(lda_result, trainY, "LDA (n=2) projection of training images of BHDD dataset")
```
